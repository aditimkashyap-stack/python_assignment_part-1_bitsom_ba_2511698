Task 1 — Data Parsing & Profile Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
cleaned_students=[]
for s in raw_students:
    name = s["name"].strip().title()
    roll = int(s["roll"])
    marks = [int(m) for m in s["marks_str"].split(", ")]

    is_valid = bool(s["name"].strip()) # Simplified validity check to ensure name is not empty
    validity = "✓ Valid name" if is_valid else "✗ Invalid name"
    cleaned_students.append({"name": name,"roll":roll,"marks":marks})
    print("================================")
    print(f"student : {name} {validity}")
    print(f"Roll no : {roll}")
    print(f"Marks : {marks}")
    print("================================")

for s in cleaned_students:
    if s["roll"] == 103:
        print (f"\nRoll 103 ALL CAPS : {s['name'].upper()}")
        print(f"Roll 103 lowercase : {s['name'].lower()}")
```

Task 2 — Marks Analysis Using Loops & Conditionals

```python
print("\n" +"="*40)
print("Task 2 : Marks Analysis")
print("="*40)

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

def get_grade(m):
  if m >=90: return "A+"
  elif m >= 80: return "A"
  elif m >= 70: return "B"
  elif m >= 60: return "C"
  else: return "F"
  print(f"\nStudent : {student_name}")
  print(f"{'Subject':<12}{'Marks':>5} Grade")
  print("-"*25)

  for subj, mark in zip(subjects, marks):
    print(f"{subj:<12}{mark:>5} {get_grade(mark)}")
    total = sum(marks)
    average = round(total/len(marks),2)
    hi_idx = marks.index(max(marks))
    lo_idx = marks.index(min(marks))
  print("f\nTotal Marks : {total}")
  print(f"Average Marks : {average}")
  print(f"Highest Marks : {max(marks)} in {subjects[hi_idx]}")
  print(f"Lowest Marks : {min(marks)} in {subjects[lo_idx]}")
```

While loop-marks entry system

```python
print("\n---Add New Subjects(type 'done' to stop)---")
new_count=0
while True:
  subj_input= input("Enter subject name(or 'done'):").strip()
  if subj_input.lower() == "done":
    break
  # The following block should be at the same indentation level as the 'if' statement
  marks_input = input(f"Enter marks for {subj_input} (0-100):").strip()
  try:
    m= int(marks_input)
    if 0 <= m <= 100:
      subjects.append(subj_input)
      marks.append(m)
      new_count += 1
    else: # This else aligns with its if
      print("Warning: Marks must be between 0 and 100. Entry skipped.") # Fixed typo
  except ValueError: # This except aligns with its try
    print("Warning: Invalid input- not a number. Entry skipped.")

# These lines are outside the while loop
updated_avg = round(sum(marks)/len(marks),2)
print(f"\nNew subjects added : {new_count}") # Fixed f-string and typo
print(f"Updated average marks : {updated_avg}")
```

Task 3 — Class Performance Summary

```python
print("\n" +"="*40)
print("Task 3 : Class Performance Summary")
print("="*40)

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print(f"\n{'Name':18}|{'Average':^7}|Status")
print("="*40)

passed, failed = 0,0
averages = []
topper_name, topper_avg = "", 0.0 # Initialize topper_avg as a float

for name, marks in class_data:
  avg = round(sum(marks)/len(marks),2)
  status = "Pass" if avg >= 60 else "Fail"
  averages.append(avg)
  if status == "Pass": passed += 1
  else: failed += 1
  if avg > topper_avg:
    topper_avg = avg # Assign only the float average
    topper_name = name # Assign the name
  # Print inside the loop to see status for each student
  print(f"{name:<18} {avg:^7} {status}")

class_avg = round(sum(averages)/len(averages),2)
print(f"\nPassed : {passed}")
print(f"Failed : {failed}")
print(f"Class topper: {topper_name} ({topper_avg})") # Corrected f-string for topper output
print(f"Class Average : {class_avg}")
```
